```python
# 回溯法
# 一、基本思想
# 1）把问题的解空间转化成了图或者树的结构表示，然后使用深度优先搜索策略进行遍历，遍历的过程中记录和寻找所有可行解或者最优解。

# 二、解题步骤（一般情况下）
# 1）针对所给题，定义问题的解空间；
# 2）确定易于搜索的解空间结构；
# 3）以深度优先方式搜索解空间，并在搜索过程中采用剪枝函数避免无效搜索。回溯法的实现方法有两种：递归和递推（也称迭代）。一般来说，一个问题两种方法都可以实现，只是在算法效率和设计复杂度上有区别。

# 三、核心名词解释
# 1）解空间树分为两种：子集树和排列树。两种在算法结构和思路上大体相同。
# 2）子集树：所给的问题是从n个元素的集合S中找出满足某种性质的子集时，相应的解空间成为子集树。
# 3）排列树：所给的问题是确定n个元素满足某种性质的排列时，相应的解空间就是排列树。
# 4）剪枝函数包括两类：使用约束函数，剪去不满足约束条件的路径；使用限界函数，剪去不能得到最优解的路径。

# 四、经典问题
# 1）装载问题
# 2）0-1背包问题
# 3）旅行售货员问题
# 4）八皇后问题
# 5）迷宫问题
# 6）图的m着色问题

# 附录：
# 1）https://blog.csdn.net/weiyuefei/article/details/79316653
# 2）https://www.cnblogs.com/riskyer/p/3279667.html
```
